Write a Python program to add an item to a tuple.






```python
tuplex = (4, 6, 2, 8, 3, 1)
print(tuplex)
```

```python
tuplex = tuplex + (9,)
print(tuplex)
```

```python
tuplex = tuplex[:5] + (15,20,25) + tuplex[:5]
print(tuplex)
```

```python
listx = list(tuplex)

listx.append(30)

tuplex = tuple(listx)

print(tuplex)
```
