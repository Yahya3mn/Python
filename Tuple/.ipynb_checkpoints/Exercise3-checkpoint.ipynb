Write a Python program to remove an item from a tuple.



```python
tuplex = "w", 3, "r", "e", "s", "o", "u", "r", "c", "e"
print(tuplex)
```

```python
tuplex = tuplex[:2] + tuplex[3:]
print(tuplex)
```

```python
listx = list(tuplex)

listx.remove("c")
```

```python
tuplex = tuple(listx)
print(tuplex)
```
