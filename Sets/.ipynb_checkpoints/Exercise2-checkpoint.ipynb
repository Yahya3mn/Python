Write a Python program to remove an item from a set if it is present in the set.



```python
num_set = set([0,1,2,3,4,5])

print("Original set elements:")
print(num_set)
```

```python
print("Remove 4 from the given set:")
num_set.discard(4)
print(num_set)
```

```python
print("Remove 5 from the given set:")
num_set.discard(5)
print(num_set)
```
